```python
%matplotlib widget
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('gfa_robustness.df',delimiter=' ')
configs = []
for run in df['run']:
    if run < 5:
        config = 'shasta'
    elif run < 10:
        config = 'hifiasm'
    elif run < 20:
        config = 'hs_shuf'
    elif run < 40:
        config = 'either'
    elif run < 203:
        config = 'peregrine'
    elif run < 206:
        config = 'raven'
    elif run < 209:
        config = 'flye'
    elif run < 210:
        config = 'hicanu'
    elif run < 220:
        config = 'all_shuf'
    elif run < 240:
        config = 'all_eith'
    configs.append(config)
df['config'] = configs
```

```python
sns.pairplot(data=df,corner=True,hue='run',diag_kind='hist')
```

```python
import scipy.stats as ss
from collections import Counter
dfq=df[df['chrom']==1]

raws = []
count= 0
for t in ['total_node', 'len_node', 'ref_node', 'ref_node_len', 'non_ref_node', 'non_ref_node_len', 'total_edges', 'total_rr', 'total_rn', 'total_nn']:#['non_ref_node_len','total_nn']:
    for j in combinations(['hicanu','peregrine','hifiasm','shasta','flye','raven','either','all_either'],2):
        try:
            val = ss.mannwhitneyu(dfq[dfq['config']==j[0]][t],dfq[dfq['config']==j[1]][t])[1]
        except:
            continue
        count += 1
        if val < 0.05:
            print('*******',j,t,val)
            raws.append(j)
print(Counter(raws))
            #print(dfq[dfq['config']==j[0]][t],dfq[dfq['config']==j[1]][t])
print(count)
    
```
