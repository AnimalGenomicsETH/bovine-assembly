```python
%matplotlib widget
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('ars.stats.df',delimiter=' ')
configs = []
for run in df1['run']:
    if run < 5:
        config = 'shasta'
    elif run < 10:
        config = 'hifiasm'
    elif run < 20:
        config = 'hs_shuf'
    elif run < 40:
        config = 'either'
    elif run < 102:
        config = 'LOW_shasta'
    elif run < 104:
        config = 'LOW_hifiasm'
    elif run < 107:
        config = 'LOW_hs_shuf'
    elif run < 110:
        config = 'LOW_either'
    elif run < 203:
        config = 'peregrine'
    elif run < 206:
        config = 'raven'
    elif run < 209:
        config = 'flye'
    elif run < 210:
        config = 'hicanu'
    elif run < 220:
        config = 'all_shuf'
    elif run < 240:
        config = 'all_eith'
    configs.append(config)
df1['config'] = configs
df1['ref']='ARS'

df2 = pd.read_csv('angus.stats.df',delimiter=' ')
df3 = pd.read_csv('simmental.stats.df',delimiter=' ')
configs = []
for run in df2['run']:
    if run < 2:
        config = 'shasta'
    elif run < 4:
        config = 'hifiasm'
    elif run < 7:
        config = 'hs_shuf'
    elif run < 10:
        config = 'either'
    configs.append(config)
df2['config']=configs
df3['config']=configs
df2['ref']='Angus'
df3['ref']='Simmental'

df = pd.concat([df1,df2,df3])
```

```python
sns.pairplot(data=df,corner=True,hue='run',diag_kind='hist')
```
