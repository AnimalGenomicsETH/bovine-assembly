```python
%matplotlib widget
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('sampling_data.three.df')

coverage = {'shasta': 55.41227207,'sp': 110.41227207,'shasta_R': 55.41227207,'hifiasm_TB': 26.384424594814814,'hifiasm_AG': 51.93792347555556,'hifiasm_AGR': 51.93792347555556,'hifiasm_R': 26.384424594814814,'hifiasm_RR': 26.384424594814814}
df['coverage'] = [coverage[asm]*s/100 for asm,s in zip(df['assembler'],df['sample'])]

df_pivot = df.pivot_table(columns='metric',values='value',index=['assembler','sample','coverage']).reset_index()
print(df_pivot.head())
print(df_pivot['size']-df_pivot['auto_size'])
```

```python
hues = {'shasta': 'r','sp': 'orange','shasta_R': 'r','hifiasm_TB': 'b','hifiasm_AG': 'g','hifiasm_AGR': 'g','hifiasm_R': 'b','hifiasm_RR': 'b'}
styles = {'shasta': 'x','sp': 110.41227207,'shasta_R': 55.41227207,'hifiasm_TB': 26.384424594814814,'hifiasm_AG': 51.93792347555556,'hifiasm_AGR': 51.93792347555556,'hifiasm_R': 26.384424594814814,'hifiasm_RR': 26.384424594814814}

d_hue = [hues[a] for a in df['assembler']]
```

```python
g=sns.relplot(data=df,kind='scatter',col='metric',x='coverage',y='value',hue='assembler',hue_order=hues.keys(),palette=hues.values(),col_wrap=4,facet_kws={'sharey':False},col_order=['NG50','P50','gaps','QV','completeness','BUSCO','masked','satellite','size','auto_size'])#,hue_order=['g','g','r','r','b','b','c','c'])
for i in [0,1,2,7]:
    g.axes[i].set_yscale('log')
plt.tight_layout()
plt.show()
```
