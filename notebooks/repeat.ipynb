```python
%matplotlib widget
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
from itertools import product
df = pd.concat([pd.read_csv(f'hap{N}_100_{S}.repeats.csv') for N,S in product((1,2),('hifiasmv152','shastamerfintigs'))])
```

```python
dfx = df[df['primary']==False]
t = dfx.groupby(['assembler','hap']).sum().reset_index()
t['chrm']='unplaced'
```

```python
df2 =pd.concat([df[df['primary']],t])
```

```python
for c in ['GC', 'Retroelements', 'SINEs', 'LINEs','LTR elements', 'DNA transposons', 'Unclassified','Total interspersed repeats', 'Small RNA', 'Satellites','Simple repeats', 'Low complexity','masked']:
    df2[c+'_n'] = df2[c]/df2['length']
df2
```

```python
sns.scatterplot(data=df2,x='chrm',y='LINEs_n',hue='assembler')
```

```python
melt = df2.melt(id_vars=['chrm','hap','assembler','primary','length'],var_name='repeat')
```

```python
sns.catplot(data=melt,x='chrm',y='value',col='repeat',col_wrap=4,hue='assembler',sharey=False,col_order=['masked_n','GC_n','SINEs_n','LINEs_n','LTR elements_n','DNA transposons_n','Unclassified_n','Satellites_n'])
```

```python
df2.groupby('assembler')[['masked','GC_n','SINEs_n','LINEs_n','LTR elements_n','DNA transposons_n','Unclassified_n','Satellites_n','Total interspersed repeats_n','Small RNA_n','Simple repeats_n', 'Low complexity_n']].mean()
```
